```python
import requests
import json
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
response = r.text
data = json.loads(response)
total = len(data)
print("回傳" + str(total)) + "筆資料")
print("\n")
print("每一筆資料包含:")
for key in data[0].keys():
    print(key)
print("\n")

print("取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」:")
for rows in data:
    print(rows['title'], rows['createdAt'], rows['commentCount'], rows['likeCount'])
print("\n")

print("計算熱門文章的「平均留言人數」與「平均按讚人數」:")
sumComment = 0
sumLike = 0
for rows in data:
    sumComment = sumComment + rows['commentCount']
    sumLike = sumLike + rows['likeCount']
print(sumComment / total , sumLike)

print("計算非熱門文章的「平均留言人數」與「平均按讚人數」:")
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')
response = r.text
data = json.loads(response)
sumComment = 0
sumLike = 0
for rows in data:
    sumComment = sumComment + rows['commentCount']
    sumLike = sumLike + rows['likeCount']
print(sumComment, sumLike)
```
